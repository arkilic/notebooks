```python
dm = pd.read_csv("~/Downloads/DM-History-20171101.csv")
price = pd.read_csv("~/Downloads/Price-History-20171101.csv")
```

```python
dm = dm.dropna()
price = price.dropna()
```

```python
dm.reset_index(drop=True, inplace=True)
price.reset_index(drop=True, inplace=True)
```

```python
dm
```

```python
# Get all dates and cusips
dates = dm['Date']
cusips = dm.columns[1:56]
```

```python
dx = pd.DataFrame(columns=['Date', 'CUSIP', 'DM', 'PRICE'])
l = 0
for i in range(len(dates)):
    for j in range(len(cusips)):
        dm_val = dm[dm['Date']==dates[i]][cusips[j]].values[0]
        price_val = price[price['Date']==dates[i]][cusips[j]].values[0]
        cusip = str(cusips[j])
        dx.loc[l] = [dates[i], cusip, dm_val, price_val]
        l += 1
```

```python
dx.head()
```

```python
dx['Date'] = dx['Date'].astype('datetime64[ns]')

dx.to_csv('/Users/armanarkilic/Desktop/historical.csv', index=False, date_format='%Y-%m-%d')
```
