```python
import pandas as pd
import numpy as np
```

```python
factors = pd.read_csv("/Users/armanarkilic/Desktop/BECM/factors.csv")
libor = pd.read_csv("/Users/armanarkilic/git/vista-pxproc/dat/REFDATA/LIBOR.csv")
spread = pd.read_csv("/Users/armanarkilic/Desktop/libor_spread.csv")
factors.head()
```

```python
dates = factors['Date']
cusips = factors.columns[1:]
```

```python
factors['Date'] = pd.to_datetime(factors['Date'])
libor['DATE'] = pd.to_datetime(libor['DATE'])
libor.index = libor.DATE
facts = factors[factors['Date'] >= '2016-12-27']
facts
```

```python
# Forward fill libor table
libor = libor.resample("D").ffill() #forward fill libor data
libor.head()
```

```python
dates = facts['Date']
cusips = facts.columns[1:]
```

```python
dx = pd.DataFrame(columns=['date', 'CUSIP','FACTOR', 'LIBOR', 'SPREAD'])
l = 0
for i in range(len(dates)):
    for j in range(len(cusips)):
        fact_val = facts[facts['Date']==dates[i]][cusips[j]].values[0]
        libor_val = libor[libor.index==dates[i]]['RATE'].values[0]
        spread_val = spread[spread.CUSIP==cusips[i]].FLT_SPREAD.values[0]
        cusip = str(cusips[j])
        dx.loc[l] = [dates[i], cusip, fact_val, libor_val, spread_val]
        l += 1
```
