```python
import sys
import csv
import dsv
import pandas as pd
from collections import OrderedDict
```

```python
with open('/Users/arkilic/Downloads/mockup_data_articles.dsv', 'rt') as csvfile:
    reader = csv.reader(csvfile, dialect=dsv.DSVDialect)
    columns = next(reader)[0].split('|')
    ddict = dict()
    data = [r[0].split('|') for r in reader]
    uids = set([d[0] for d in data])
    uids = list(uids)
    colnames = ['ARTICLE_NO']
    partnos = set([int(d[1]) for d in data])
    colnames.extend(partnos)
    # init data frame
    df = pd.DataFrame(data={"ARTICLE_NO": uids
                                }, columns=colnames)
    df.set_index('ARTICLE_NO')
```

```python
for d in data:
    rowidx = df[df['ARTICLE_NO'] == d[0]].index.tolist()[0]
    # explicitly name the cols for more readable code
    partid = int(d[1])
    partcol = d[2]
    df.loc[rowidx][partid] = partcol
```
